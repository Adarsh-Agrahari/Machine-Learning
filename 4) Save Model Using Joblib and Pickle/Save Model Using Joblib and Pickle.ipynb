```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
```

```python
model = linear_model.LinearRegression()
model.fit(df[['area']],df.price)
```

```python
model.predict([[3300]])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[5000]])
```

```python
with open('model_pickle','wb') as f:
    pickle.dump(model,f)
```
